```python
# Standard imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_percentage_error, \
  mean_absolute_error, r2_score, mean_squared_error

df = pd.read_pickle('/content/drive/Othercomputers/My MacBook Pro/KJY/Data Science/Data_Scientist_Position_at_Etiqa_Insurance_Singapore/0-df.pkl')
```

```python
df
```

```python
# naive forecast - use y(t) to predict y(t+1)
df['calls_prediction'] = df['calls'].shift(1)
```

```python
df
```

```python
y_true = df.iloc[1:]['calls']
y_pred = df.iloc[1:]['calls_prediction']
```

# Metrics
- MSE
- RMSE
- R^2

```python
# MSE
mean_squared_error(y_true, y_pred)
```

```python
# MSE function
def calculate_mse(x, y):
  return (x - y).dot(x - y) / len(x)

calculate_mse(y_true, y_pred)
```

```python
# RMSE
mean_squared_error(y_true, y_pred, squared=False)
```

```python
# RMSE function
def calculate_rmse(x, y):
  return np.sqrt((x - y).dot(x - y) / len(x))

calculate_rmse(y_true, y_pred)
```

```python
# R^2
r2_score(y_true, y_pred)
```
